```python
import os
import pandas as pd
import glob
import csv

# Define the directory containing the input files and the output directory
input_directory = '.'
output_directory = 'csv/'

# Ensure the output directory exists
os.makedirs(output_directory, exist_ok=True)

# Get a list of all files in the input directory with a specific extension (e.g., .txt)
input_files = glob.glob(os.path.join(input_directory, '*.pf'))

```

```python
input_files
```

```python

# Process each file
for input_file in input_files:
    # Define the output CSV file path
    base_filename = os.path.basename(input_file)
    output_csv = os.path.join(output_directory, os.path.splitext(base_filename)[0] + '.csv')
    
    # Read the data from the input file using pandas
    with open(input_file, 'r') as file:
        lines = file.readlines()
    
    # Write the data to the CSV file
    with open(output_csv, 'w', newline='') as csvfile:
        csv_writer = csv.writer(csvfile)
        
        for line in lines:
            # Split the line into values based on tab character
            row = line.strip().split('\t')
            if len(row) == 1:
                row = line.strip().split(' ')
            # Write the row to the CSV file
            csv_writer.writerow(row)
    
    print(f'{input_file} has been successfully transformed into {output_csv}')
```
